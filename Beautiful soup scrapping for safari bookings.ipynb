```python
url="https://www.safaribookings.com/reviews/p2346"
data=requests.get(url)
soup=BeautifulSoup(data.content,'html.parser')
print(soup)
```

```python
def cus_name(soup):
    # find the Html tag
    # with find()
    # and convert into string
    data_str = ""
    cus_list = []
  
    for item in soup.find_all("strong", class_="review__person__name"):
        data_str = data_str + item.get_text()
        cus_list.append(data_str)
        data_str = ""
    return cus_list
  
  
cus_res= cus_name(soup)
print(cus_res)
```
